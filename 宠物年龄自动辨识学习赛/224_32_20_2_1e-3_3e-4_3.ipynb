```python
import os
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
import shutil
from datetime import datetime
from zoneinfo import ZoneInfo
import numpy as np
import timm

# --- 全局常量与定义 ---
ANNOTATION_PATH = '/kaggle/input/cleandata/clean/annotations/'
TRAIN_IMG_PATH = '/kaggle/input/cleandata/clean/trainset/'
VAL_IMG_PATH = '/kaggle/input/cleandata/clean/valset/'
OUTPUT_PATH = '/kaggle/working/'
WEIGHTS_PATH = '/kaggle/input/bestmodel2/pytorch/default/1/best_model.pth'
IMG_SIZE = 224
BATCH_SIZE = 32
NUM_EPOCHS = 20
WARMUP_EPOCHS = 2
L2_LAMBDA = 1e-3
BASE_LR = 3e-4
PATIENCE = 3

# --- 函数与类的定义 ---
def parse_txt_to_df(txt_path):
    """从 txt 文件解析数据到 DataFrame"""
    uuids = []
    ages = []
    try:
        with open(txt_path, 'r') as f:
            for line in f.readlines():
                line = line.strip()
                if line:
                    parts = line.split()
                    uuids.append(parts[0])
                    ages.append(int(parts[-1]))
    except FileNotFoundError:
        print(f"错误: 未找到文件 {txt_path}")
        return pd.DataFrame({'uuid': [], 'age': []})
    df = pd.DataFrame({'uuid': uuids, 'age': ages})
    return df

class PetAgeDataset(Dataset):
    """自定义的 PyTorch 数据集类"""
    def __init__(self, df, img_dir, transform=None):
        self.df = df
        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_name = self.df.loc[idx, 'uuid']
        img_path = os.path.join(self.img_dir, img_name)
        try:
            image = Image.open(img_path).convert('RGB')
        except FileNotFoundError:
            print(f"警告: 图像 {img_path} 未找到，跳过")
            return None, None
        label = self.df.loc[idx, 'age']
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.float32)

class WeightedL1Loss(nn.Module):
    """加权 L1 损失，加入标签平滑和 L2 正则化"""
    def __init__(self, weights, smoothing=0.1, l2_lambda=0.0):
        super(WeightedL1Loss, self).__init__()
        self.weights = weights
        self.smoothing = smoothing
        self.l2_lambda = l2_lambda

    def forward(self, outputs, targets, model):
        weights = self.weights[targets.long()]
        smoothed_targets = targets * (1 - self.smoothing) + self.smoothing * targets.mean()
        l1_loss = torch.mean(torch.abs(outputs - smoothed_targets) * weights)
        if self.l2_lambda > 0:
            l2_loss = 0
            for param in model.parameters():
                if param.requires_grad:
                    l2_loss += torch.norm(param, p=2) ** 2
            l1_loss += self.l2_lambda * l2_loss
        return l1_loss

def linear_warmup_scheduler(optimizer, epoch, warmup_epochs, base_lr):
    """线性学习率预热"""
    if epoch < warmup_epochs:
        lr = base_lr * (epoch + 1) / warmup_epochs
        for param_group in optimizer.param_groups:
            param_group['lr'] = lr
    return optimizer.param_groups[0]['lr']

def count_trainable_parameters(model):
    """计算模型中可训练参数的数量"""
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def archive_experiment_results(final_mae, optimizer, image_size, batch_size, epochs_trained):
    """归档实验结果"""
    print("\n--- 开始归档本次训练成果 ---")
    learning_rate = optimizer.param_groups[0]['lr']
    tz = ZoneInfo("Asia/Singapore")
    current_time = datetime.now(tz)
    folder_name = current_time.strftime("%Y-%m-%d_%H-%M-%S")
    experiment_path = os.path.join(OUTPUT_PATH, folder_name)
    os.makedirs(experiment_path, exist_ok=True)
    print(f"成功创建归档文件夹: {experiment_path}")

    summary_text = f"""
========================================
      宠物年龄识别模型 - 训练总结
========================================
归档时间: {current_time.strftime("%Y-%m-%d %H:%M:%S %Z")}
--- 训练参数 ---
  - 图像尺寸 (Image Size): {image_size} x {image_size}
  - 批次大小 (Batch Size): {batch_size}
  - 基础学习率 (Base Learning Rate): {BASE_LR}
  - 预热周期 (Warmup Epochs): {WARMUP_EPOCHS}
  - 训练轮数 (Epochs): {epochs_trained}
  - 模型架构 (Architecture): EfficientNet-B0
--- 最终结果 ---
  - 最佳验证集平均绝对误差 (Best Validation MAE): {final_mae:.4f} 月
========================================
"""
    summary_file_path = os.path.join(experiment_path, 'summary.txt')
    model_save_path = os.path.join(experiment_path, 'best_model.pth')
    image_save_path = os.path.join(experiment_path, 'prediction_showcase.png')
    error_dist_path = os.path.join(experiment_path, 'error_distribution.png')
    mae_curve_path = os.path.join(experiment_path, 'mae_curve.png')
    pred_true_path = os.path.join(experiment_path, 'pred_vs_true.png')
    grad_norm_path = os.path.join(experiment_path, 'grad_norm.png')
    error_bin_path = os.path.join(experiment_path, 'error_by_age_bin.csv')
    tta_variance_path = os.path.join(experiment_path, 'tta_variance.csv')

    with open(summary_file_path, 'w', encoding='utf-8') as f:
        f.write(summary_text)
    print(f"训练总结已保存至: {summary_file_path}")
    try:
        shutil.copy(os.path.join(OUTPUT_PATH, 'best_model.pth'), model_save_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'prediction_showcase.png'), image_save_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'error_distribution.png'), error_dist_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'mae_curve.png'), mae_curve_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'pred_vs_true.png'), pred_true_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'grad_norm.png'), grad_norm_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'error_by_age_bin.csv'), error_bin_path)
        shutil.copy(os.path.join(OUTPUT_PATH, 'tta_variance.csv'), tta_variance_path)
        print("模型文件、结果图、误差分布图、MAE曲线、预测-真实图、梯度范数图、年龄分段误差和TTA方差已成功备份。")
    except FileNotFoundError as e:
        print(f"归档失败: {e}")
    print("\n🎉 归档完成！所有成果已保存至以下 Kaggle 输出目录:")
    print(f"   {experiment_path}")

def main():
    """主执行函数"""
    print("步骤 1: 项目设置...")
    train_transform = transforms.Compose([
        transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    val_transform = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    print("步骤 2: 解析并加载数据...")
    train_df = parse_txt_to_df(os.path.join(ANNOTATION_PATH, 'train.txt'))
    val_df = parse_txt_to_df(os.path.join(ANNOTATION_PATH, 'val.txt'))
    
    if train_df.empty or val_df.empty:
        print("错误: 训练集或验证集为空，程序退出")
        return
    
    before_train = len(train_df)
    train_df = train_df[train_df['age'] < 192].reset_index(drop=True)
    removed_train = before_train - len(train_df)
    print(f"  - 已过滤训练集噪声样本: {removed_train} / {before_train} (≈{removed_train/before_train:.1%})")
    
    before_val = len(val_df)
    val_df = val_df[val_df['age'] < 192].reset_index(drop=True)
    removed_val = before_val - len(val_df)
    print(f"  - 已过滤验证集噪声样本: {removed_val} / {before_val} (≈{removed_val/before_val:.1%})")
    print(f"  - 找到 {len(train_df)} 个训练样本。")
    print(f"  - 找到 {len(val_df)} 个验证样本。")
     
    print("训练集年龄分布：\n", train_df['age'].describe())
    print("验证集年龄分布：\n", val_df['age'].describe())

    age_counts = train_df['age'].value_counts().sort_index()
    max_age = int(max(train_df['age'].max(), val_df['age'].max())) + 1
    weights = torch.ones(max_age, dtype=torch.float32)
    for age in age_counts.index:
        weights[age] = min(2.0 / (age_counts[age] / len(train_df)), 20.0)
    weights = weights / weights.sum() * max_age

    train_dataset = PetAgeDataset(train_df, TRAIN_IMG_PATH, transform=train_transform)
    val_dataset = PetAgeDataset(val_df, VAL_IMG_PATH, transform=val_transform)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=1, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=1)

    print("步骤 3: 定义模型...")
    # Try loading EfficientNet-B0 with pretrained weights, fall back to random initialization
    try:
        model = timm.create_model('efficientnet_b0', pretrained=True)
        print("  - 成功加载预训练权重从 Hugging Face")
    except Exception as e:
        print(f"  - 无法加载预训练权重: {e}. 使用随机初始化权重...")
        model = timm.create_model('efficientnet_b0', pretrained=False)

    num_ftrs = model.classifier.in_features
    model.classifier = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(num_ftrs, 1)
    )
    # Try loading local weights if available
    try:
        state_dict = torch.load(WEIGHTS_PATH, map_location=torch.device('cpu'))
        model.load_state_dict(state_dict, strict=False)
        print(f"  - 成功加载本地权重文件: {WEIGHTS_PATH}")
        print(f"  - 加载的 classifier 层参数: {[k for k in state_dict.keys() if 'classifier' in k]}")
    except Exception as e:
        print(f"  - 加载本地权重失败: {e}. 将继续使用当前权重训练...")

    for name, param in model.named_parameters():
        if 'blocks.4' not in name and 'blocks.5' not in name and 'blocks.6' not in name and 'classifier' not in name:
            param.requires_grad = False
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    print(f"  - 模型已移至: {device}")
    print(f"  - 可训练参数数量: {count_trainable_parameters(model)}")

    print("步骤 4: 开始训练...")
    criterion = WeightedL1Loss(weights.to(device), smoothing=0.05, l2_lambda=L2_LAMBDA)
    optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=BASE_LR, weight_decay=1e-2)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS - WARMUP_EPOCHS, eta_min=1e-6)
    train_mae_history = []
    val_mae_history = []
    grad_norms = []
    tta_variances = []
    best_val_mae = float('inf')
    best_epoch = 0
    early_stop_counter = 0

    for epoch in range(NUM_EPOCHS):
        current_lr = linear_warmup_scheduler(optimizer, epoch, WARMUP_EPOCHS, base_lr=BASE_LR)
        
        model.train()
        train_loop = tqdm(train_loader, desc=f"训练 Epoch [{epoch+1}/{NUM_EPOCHS}]")
        running_loss = 0.0
        train_mae = 0.0
        for images, labels in train_loop:
            if images is None or labels is None:
                continue
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images).squeeze()
            outputs = torch.clamp(outputs, 0, max_age)
            loss = criterion(outputs, labels, model)
            loss.backward()
            grad_norm = torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0).item()
            optimizer.step()
            running_loss += loss.item() * images.size(0)
            train_mae += torch.abs(outputs - labels).sum().item()
            train_loop.set_postfix(loss=loss.item(), lr=current_lr, grad_norm=grad_norm)

        train_epoch_loss = running_loss / len(train_dataset)
        train_epoch_mae = train_mae / len(train_dataset)
        grad_norms.append(grad_norm)

        model.eval()
        val_loop = tqdm(val_loader, desc=f"验证 Epoch [{epoch+1}/{NUM_EPOCHS}]")
        val_mae = 0.0
        val_predictions = []
        val_labels = []
        batch_tta_variances = []
        with torch.no_grad():
            for images, labels in val_loop:
                if images is None or labels is None:
                    continue
                images, labels = images.to(device), labels.to(device)
                outputs_orig = model(images).squeeze()
                outputs_flip = model(torch.flip(images, dims=[3])).squeeze()
                outputs = (outputs_orig + outputs_flip) / 2.0
                outputs = torch.clamp(outputs, 0, max_age)
                batch_variance = torch.var(torch.stack([outputs_orig, outputs_flip]), dim=0).mean().item()
                batch_tta_variances.append(batch_variance)
                val_mae += torch.abs(outputs - labels).sum().item()
                val_predictions.extend(outputs.cpu().numpy())
                val_labels.extend(labels.cpu().numpy())

        val_epoch_mae = val_mae / len(val_dataset)
        avg_tta_variance = sum(batch_tta_variances) / len(batch_tta_variances) if batch_tta_variances else 0
        tta_variances.append(avg_tta_variance)
        if val_epoch_mae == 0:
            print("警告: 验证MAE为0，可能存在预测问题")
        train_mae_history.append(train_epoch_mae)
        val_mae_history.append(val_epoch_mae)
        print(f"Epoch [{epoch+1}/{NUM_EPOCHS}] -> 训练损失: {train_epoch_loss:.4f}, 训练MAE: {train_epoch_mae:.4f}, 验证MAE: {val_epoch_mae:.4f}, 学习率: {current_lr:.6f}, 梯度范数: {grad_norm:.4f}, TTA方差: {avg_tta_variance:.4f}")

        timestamp = datetime.now(ZoneInfo("Asia/Singapore")).strftime("%Y%m%d_%H%M%S")
        torch.save(model.state_dict(), os.path.join(OUTPUT_PATH, f'model_epoch_{epoch+1}_{timestamp}.pth'))
        if val_epoch_mae < best_val_mae:
            best_val_mae = val_epoch_mae
            best_epoch = epoch + 1
            shutil.copy(os.path.join(OUTPUT_PATH, f'model_epoch_{epoch+1}_{timestamp}.pth'), os.path.join(OUTPUT_PATH, 'best_model.pth'))
            print(f"  - 验证集MAE提升至 {val_epoch_mae:.4f}。保存最佳模型: best_model.pth")
            early_stop_counter = 0
        else:
            early_stop_counter += 1

        if early_stop_counter >= PATIENCE:
            print(f"早停触发: 验证集MAE在 {PATIENCE} 个 epoch 内未提升。停止训练。")
            break

        if epoch >= WARMUP_EPOCHS:
            scheduler.step()

    print(f"\n训练完成。最佳验证集MAE: {best_val_mae:.4f} (Epoch {best_epoch})")

    experiment_path = os.path.join(OUTPUT_PATH, datetime.now(ZoneInfo("Asia/Singapore")).strftime("%Y-%m-%d_%H-%M-%S"))
    os.makedirs(experiment_path, exist_ok=True)

    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(train_mae_history) + 1), train_mae_history, label='Train MAE')
    plt.plot(range(1, len(val_mae_history) + 1), val_mae_history, label='Validation MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE (months)')
    plt.legend()
    plt.savefig(os.path.join(experiment_path, 'mae_curve.png'))
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(grad_norms) + 1), grad_norms, label='Gradient Norm')
    plt.xlabel('Epoch')
    plt.ylabel('Gradient Norm')
    plt.legend()
    plt.savefig(os.path.join(experiment_path, 'grad_norm.png'))
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.scatter(val_labels, val_predictions, alpha=0.5)
    plt.plot([min(val_labels), max(val_labels)], [min(val_labels), max(val_labels)], 'r--')
    plt.xlabel('True Age (months)')
    plt.ylabel('Predicted Age (months)')
    plt.title('Predicted vs True Age')
    plt.savefig(os.path.join(experiment_path, 'pred_vs_true.png'))
    plt.show()

    errors = [abs(pred - true) for pred, true in zip(val_predictions, val_labels)]
    plt.hist(errors, bins=30)
    plt.xlabel('Absolute Error (months)')
    plt.ylabel('Count')
    plt.savefig(os.path.join(experiment_path, 'error_distribution.png'))
    plt.show()
    print("验证集误差分布统计：\n", pd.Series(errors).describe())

    tta_stats = pd.Series(tta_variances)
    tta_stats.to_csv(os.path.join(experiment_path, 'tta_variance.csv'), header=['variance'], index_label='epoch')
    print("TTA方差统计：\n", tta_stats.describe())

    val_df_pred = pd.DataFrame({'true_age': val_labels, 'pred_age': val_predictions})
    bins = pd.qcut(val_df_pred['true_age'], q=4, duplicates='drop')
    val_df_pred['age_bin'] = bins
    error_by_bin = val_df_pred.groupby('age_bin', observed=True).apply(
        lambda x: pd.Series({
            'mean_error': (x['pred_age'] - x['true_age']).abs().mean(),
            'count': len(x)
        }), include_groups=False
    )
    print("按年龄分段的平均误差：\n", error_by_bin)
    error_by_bin.to_csv(os.path.join(experiment_path, 'error_by_age_bin.csv'), encoding='utf-8')

    print("\n步骤 5: 最终评估与结果展示...")
    try:
        model.load_state_dict(torch.load(os.path.join(experiment_path, 'best_model.pth')))
        print(f"  - 成功加载最佳模型: best_model.pth")
    except FileNotFoundError:
        print(f"  - 未找到最佳模型文件: best_model.pth，尝试加载最后模型")
        try:
            model.load_state_dict(torch.load(os.path.join(experiment_path, f'model_epoch_{best_epoch}_{timestamp}.pth')))
            print(f"  - 成功加载最后模型: model_epoch_{best_epoch}_{timestamp}.pth")
        except FileNotFoundError:
            print(f"  - 未找到任何模型文件，跳过加载")
    model.to(device)
    model.eval()

    def visualize_predictions(dataset, num_samples=5):
        inv_normalize = transforms.Normalize(
            mean=[-0.485/0.229, -0.456/0.224, -0.406/0.225],
            std=[1/0.229, 1/0.224, 1/0.225]
        )
        plt.figure(figsize=(15, 5 * num_samples))
        indices = random.sample(range(len(dataset)), num_samples)
        for i, idx in enumerate(indices):
            image, label = dataset[idx]
            if image is None or label is None:
                continue
            img_for_display = inv_normalize(image).permute(1, 2, 0)
            img_for_display = (img_for_display * 255).clamp(0, 255).byte().numpy()
            with torch.no_grad():
                image_tensor = image.unsqueeze(0).to(device)
                prediction = model(image_tensor).squeeze().item()
            plt.subplot(num_samples, 1, i + 1)
            plt.imshow(img_for_display)
            plt.title(f"sample {i+1}: real age = {label:.1f} month, pre_age = {prediction:.1f} month")
            plt.axis('off')
        plt.tight_layout()
        plt.savefig(os.path.join(experiment_path, 'prediction_showcase.png'))
        plt.show()

    visualize_predictions(val_dataset)
    print("\n程序结束！请查看生成的图表以及 'prediction_showcase.png', 'error_distribution.png', 'mae_curve.png', 'pred_vs_true.png', 'grad_norm.png', 'error_by_age_bin.csv' 和 'tta_variance.csv' 文件。")

    archive_experiment_results(
        final_mae=best_val_mae,
        optimizer=optimizer,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        epochs_trained=epoch + 1
    )

if __name__ == '__main__':
    main()
```
